```python
# Ecrire un programme qui calcule une r´ealisation d’un mouvement Brownien Wt, 
# sur un horizon T = 1 avec N = 500, h = T/N
```

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
N=500
T=1
t=np.linspace(0,T,N+1,endpoint=True)
h=T/N

#Simulation des accroissements 

dW = np.random.normal(0,h,N)

#Simulation d'une trajectoire 

W = np.zeros(N+1)
W[1:]=np.cumsum(dW)

plt.plot(t,W)
plt.grid()
plt.show()

W[0]
```
